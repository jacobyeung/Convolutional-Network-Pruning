```python
import matplotlib.pyplot as plt
import os, glob
```

```python
se = 303
model_dir = f'outputs/experiment_{se}/data'
model_dirs = list(glob.glob(os.path.join(model_dir, '*0.*')))
```

```python
model_dirs = [x for x in model_dirs if x[-2:] != 'l1']
```

```python
model_dirs
```

```python
model_names = [list(glob.glob(os.path.join(x, '*.pt'))) for x in model_dirs]
```

```python
valid_accs = []
for per in model_names:
    temp = []
    for f in per:
        direc, model = f.split('/')[-2:]
#         perc = direc.split('.')[-1][:-3]
        perc = direc.split('.')[-1]
        percent = f'0.{perc}'
        iteration = model.split('_')[1]
        acc = model.split('=')[-1][:-3]
        temp.append((float(percent), int(iteration), float(acc)))
    temp.sort()
    valid_accs.append(temp)
```

```python
valid_accs.sort()
valid_accs
```

```python
accs = []
for i in range(3):
    tempx = []
    tempy = []
    for j in range(len(valid_accs)):
        tempx.append(valid_accs[j][i][0])
        tempy.append(valid_accs[j][i][2])
    accs.append((tempx, tempy))
```

```python
markers = ['o', '*', 'x']
labels = ['Validation 1', 'Validation 2', 'Validation 3']
for i in range(3):
    plt.plot(accs[i][0], accs[i][1], marker=markers[i], label=labels[i])
plt.legend()
tuck = 'Rank-1 Tucker Decomposition'
l1 = 'L1'
method = tuck
plt.title(f"Pruning Filters Using {method}")
plt.xlabel('Percentage of Filters Pruned')
plt.ylabel('Validation Accuracy')
plt.savefig(f"prune vs acc {method}.png")
```
